```python
#Estimating the total compensation to be provided to an employee
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean
```

```python
df=pd.read_csv('train_set.csv')
df
```

```python
 df.shape
```

```python
features = df[['Salaries', 'Overtime', 'H/D']]  
target = df['Total_Compensation']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
print('Root Mean Squared Error:', rmse)
```

```python
all_predictions = model.predict(features)
df['Estimated_Total_Compensation'] = all_predictions
```
